```python
from dgsp.estimators import TrivialEstimator
```

```python
import numpy as np
```

```python
all_traj = []
for i in range(10000):
    tensor = np.load(f'../data/traj/{i}.npy')
    all_traj.append(tensor)
    if i % 100 == 0:
        print(i // 100)
all_traj = np.array(all_traj)
```

```python
est = TrivialEstimator(all_traj)
```

```python
traj = np.load('../data/traj/0.npy')
```

```python
import matplotlib.pyplot as plt
```

```python
for i in range(4):
    plt.plot(traj[:, i])
    plt.plot(est.traj[:, i])
    plt.show()
```
