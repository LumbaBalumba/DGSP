```python
from dgsp.estimators import ParticleFilter
```

```python
filter = ParticleFilter(dt=1e-3, n_particles=1000)
```

```python
filter.particles.shape
```

```python
filter.weights.shape
```

```python
import numpy as np
```

```python
np.average(filter.particles, weights=filter.weights, axis=0)
```

```python
for i in range(1000):
    filter.predict()
    if i % 100 == 0:
        filter.update(np.zeros(2))
```
